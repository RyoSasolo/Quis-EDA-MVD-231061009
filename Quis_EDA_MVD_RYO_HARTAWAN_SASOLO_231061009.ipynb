*Fungsi Mengganti Nol dengan Median*

def replace_zeros_with_median(df, column_name):
    """Mengganti nilai 0 dalam kolom dengan median kolom."""
    median_value = df[column_name][df[column_name] != 0].median()
    df[column_name] = df[column_name].replace(0, median_value)
    return df

```
data = data.replace('', pd.NA)

data['Customer Value'] = pd.to_numeric(data['Customer Value'], errors='coerce')

def replace_zeros_with_median(df, column_name):
    """Mengganti nilai 0 dalam kolom dengan median kolom."""
    median_value = df[column_name][df[column_name] != 0].median()
    df[column_name] = df[column_name].replace(0, median_value)
    return df

numerical_columns = ['Call Failure', 'Charge Amount', 'Seconds of Use', 'Freq. of use', 'Freq. of SMS', 'Distinct Called Numbers', 'Customer Value']

for column in numerical_columns:
    data = replace_zeros_with_median(data, column)

data.isnull().sum()
data
```
